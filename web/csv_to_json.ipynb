```python
import json
import pandas as pd
```

```python
translators_cost = pd.read_csv("../data/translators.csv")
translators_cost.head()
translators_sch = pd.read_csv("../data/schedules.csv")
translators_sch.head()
```

```python
hundred_trans= translators_cost[:101]
hundred_trans
```

```python
diccionario_trans = {}
# Iterar sobre las filas del DataFrame
for index, row in translators_cost.iterrows():
    # Obtener el nombre del traductor, los lenguajes y el costo
    traductor = row['TRANSLATOR']
    source_language = row['SOURCE_LANG']
    target_language = row['TARGET_LANG']
    cost = row['HOURLY_RATE']
    
    # Si el traductor no estÃ¡ en el diccionario, agregarlo con un valor vacÃ­o
    if traductor not in diccionario_trans:
        diccionario_trans[traductor] = []
    
    # AÃ±adir los datos del traductor como una tupla (lenguaje de origen, lenguaje de destino, costo)
    diccionario_trans[traductor].append({
        'Source Language': source_language,
        'Target Language': target_language,
        'Hourly_rate': cost
    })

diccionario_trans
```

```python
#provisionales
hours_available = 10
quality = 8
finish_day= "5th March"
done_in = "5h"
cost = "25 eurazos"
```

```python
json_estructura = {"translators": []}

for traductor, tareas in diccionario_trans.items():

    known_languages = set()
    for tarea in tareas:
        known_languages.add(tarea["Source Language"])
        known_languages.add(tarea["Target Language"])

    json_estructura["translators"].append({
        "name": traductor,
        "languages": tareas, # Lista con los idiomas y costos
        "hours_available": hours_available,
        "quality": quality,
        "finish_day": finish_day,
        "done_in": done_in,
        "cost": cost,
        "avatar":"/Imagenes/perfilavatar.png" 
        })

# Convertir a JSON con indentaciÃ³n para que sea mÃ¡s legible
json_result = json.dumps(json_estructura, indent=4)

# Mostrar el JSON resultante
print(json_result)
```

```python
with open("Transl_csvtojson.json", "w") as archivo:
    json.dump(json_estructura, archivo, indent=4)
```
